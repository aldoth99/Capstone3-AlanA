## **Matrix Evaluation**

Metrik evaluasi yang akan digunakan adalah:

- R-Square:

Metrik ini mewakili persentase dari semua fitur yang mempengaruhi target. Semakin tinggi nilai persentase, semakin besar pengaruh fitur terhadap target. Metrik ini hanya akan digunakan saat memilih model dasar karena dari setiap model dapat dilihat dengan jelas seberapa besar persentase pengaruh fitur dari setiap model terhadap target. Karena itu adalah persentase, nilai metrik adalah 0 hingga 1.

- MAE (Mean Absolute Error)

Metrik ini mewakili perbedaan rata-rata absolut antara hasil yang diprediksi dan hasil aktual. Metrik ini disarankan karena data yang digunakan adalah data yang memiliki beberapa outliers. Karena nilai absolut, metrik ini tidak peduli tentang perbedaan yang menghasilkan hasil negatif. Semakin kecil nilai metrik ini, semakin baik hasil prediksi yang didapat

- MAPE (Mean Absolute Percentage Error)

Metrik ini mewakili persentase perbedaan rata-rata absolut antara hasil yang diprediksi dan hasil aktual. Bedanya dengan MAE, metrik ini perlu mengubah hasil perbedaan menjadi persentase saja. Karena itu adalah persentase, nilai metrik adalah 0 hingga 1.

- RMSE (Root Mean Squared Error) 

RMSE adalah metrik evaluasi yang mewakili rata-rata kuadrat dari perbedaan antara hasil prediksi dan hasil aktual. Nilai RMSE didapat dari pembulatan kuadrat dari error sehingga memperlihatkan seberapa besar jarak antara hasil prediksi dan hasil aktual. Semakin kecil nilai RMSE, semakin baik hasil prediksi yang didapat. Namun, RMSE sangat sensitif terhadap data outliers sehingga tidak cocok untuk digunakan pada data yang memiliki beberapa outliers.

## **3.5. Outliers Detection**

Jumlah outliers pada data pasti akan mempengaruhi hasil analisis dan pemodelan yang dilakukan. Ada beberapa hal yang harus diperhitungkan sebelum memutuskan bagaimana cara menanganinya, termasuk mempertimbangkan pengetahuan tentang domain.

```python
def check_outlier(col):
    
    Q1 = np.percentile(col, 25)
    Q3 = np.percentile(col, 75)

    IQR = Q3 - Q1

    batas_bawah = Q1 - (IQR * 1.5)
    batas_atas = Q3 + (IQR * 1.5)

    data_outliers = col[(col < batas_bawah) | (col > batas_atas)]
    return data_outliers
```

```python
feature_outlier = []
for col in df.select_dtypes(['float','int']):
    outliers = check_outlier(df[col])
    if len(outliers):
        print(f"Feature",col, "memiliki", len(outliers), "outlier")
        feature_outlier.append(col)
    else:
        print(f"Feature", col, "has no outlier")
```

```python
fig, axs = plt.subplots(2, 2, figsize = (25,10))
i = 0
loca = [axs[0,0], axs[0,1], axs[1,0], axs[1,1]]

for col in feature_outlier:
    sns.boxplot(df[col], ax = loca[i], color= 'green')
    i += 1
    
plt.show()
```

***Dari gambar di atas, dapat dilihat fitur numerik memiliki banyak outlier. Dalam project ini, untuk membuat dataset yang lebih bersih dari outliers akan dilakukan beberapa manipulasi data pada fitur-fitur dengan mempertimbangkan kombinasi hasil dari metode IQR dan juga pengetahuan tentang domain. Model yang dihasilkan hanya akan dapat memprediksi harga mobil bekas yang nilainya berada pada rentang data yang bersih, di luar rentang tersebut, hasil prediksi dari model akan cenderung salah.***

```python
# Def cari outliers
def cari_outliers(data):
    q1 = data.describe()['25%']
    q3 = data.describe()['75%']
    iqr = q3 - q1
    upper_fence = q3 + (1.5*iqr)
    lower_fence = q1 - (1.5*iqr)
    print(f'limit bawah: {lower_fence}')
    print(f'limit atas: {upper_fence}')
```

```python
# outliers pada kolom price
print(f'Outlier Price')
cari_outliers(df_model['Price'])
print()


# outliers pada kolom Engine_Size
print(f'Outlier Engine Size')
cari_outliers(df_model['Engine_Size'])
print()


# outliers pada kolom Mileage
print(f'Outlier Mileage')
cari_outliers(df_model['Mileage'])
print()


# Outliers pada kolom Year
print(f'Outlier Year')
cari_outliers(df_model['Year'])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    vif = pd.DataFrame()
    vif['variables'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```

```python
# Import library untuk modeling

from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV, GridSearchCV, KFold

import category_encoders as ce
from sklearn.preprocessing import OneHotEncoder, KBinsDiscretizer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.compose import TransformedTargetRegressor

from sklearn.preprocessing import StandardScaler

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, accuracy_score
```

## **4.1. Encoding**

Untuk mengaplikasikan fitur yang dipilih, tipe data fitur kategorikal harus dikonversi menjadi numerikal menggunakan teknik encoding. Dalam hal ini, Column Transformer akan digunakan untuk melakukan konversi ini dan akan diterapkan dalam pipeline model. Fitur Options, Origin, dan Gear type akan diencode dengan onehot encoding, sementara fitur Type, Region, dan Make akan diencode dengan binary encoding. Dan fitur Mileage akan dibinning menggunakan equal frequency.

```python
# mengubah fitur kategorikal menjadi numerikal

transformer = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), ['Options','Origin','Gear_Type']),
    ('binary', ce.BinaryEncoder(), ['Type','Region','Make']),
    ('binner', KBinsDiscretizer(n_bins=4, encode='onehot', strategy='quantile'), ['Mileage'])
], remainder='passthrough')
```

## **4.3. Model Bencmarking**

Langkah berikutnya, kita akan mencari model untuk dataset. Pencarian model dibagi menjadi 3 tahap. Setelah memperoleh model dasar, tahap kedua akan melanjutkan dengan tuning hiperparameter untuk mengetahui parameter mana yang dianggap terbaik berdasarkan evaluasi metrik. Setelah mendapatkan model dasar dengan benchmark terbaik, tahap terakhir adalah mencoba model pada data uji dengan melihat hasil evaluasi metrik. ### **4.3.1 Pilih Base Model**

Berikut kita akan mencari model dasar menggunakan model yang disebutkan pada bagian "Modeling". Setiap model akan diuji pada data train dari dataset dan melihat bagaimana nilai setiap penilaian metrik-nya. Ada satu fungsi tambahan yaitu TransformedTargetRegressor. Fungsi ini membuat skala variabel Target menjadi lebih kecil sehingga data menjadi lebih normal terdistribusi. Setelah Target diproses, fungsi ini akan mengembalikan nilai awal Target ke nilai aslinya sehingga hasil yang didapat akan maksimal.

**Data Train**

```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=2020)
gdt = GradientBoostingRegressor(random_state = 2020)
rf = RandomForestRegressor(random_state=2020)
xgb = XGBRegressor(random_state=2020)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_gdt = TransformedTargetRegressor(gdt, func = np.log, inverse_func = np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

# Kandidat algoritma
models = [log_lr, log_knn, log_dt, log_gdt, log_rf, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari algoritma terbaik berdasarkan nilai MAE
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=2020)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        x_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        x_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        x_train, 
        y_train,
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error'
        )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

```python
# Evaluasi hasil dari 5 kandidat algoritma yang digunakan
pd.DataFrame({
    'Model': ['Linear Regression', 'KNN Regressor' ,'DecisionTree Regressor', 'GradientBoostingRegressor', 'RandomForest Regressor', 'XGBoost Regressor'],
    'Mean_RMSE': nilai_mean_rmse, 'Std_RMSE': nilai_std_rmse,
    'Mean_MAE': nilai_mean_mae, 'Std_MAE': nilai_std_mae,
    'Mean_MAPE': nilai_mean_mape, 'Std_MAPE': nilai_std_mape})
```

**Dari tabel di atas, dapat dilihat bahwa Rndom Forest Regressor adalah model terbaik ketika dites pada data train. Hasil dari evaluasi metrik Random Forest Regressor menunjukkan bahwa setiap parameter metrik memiliki hasil terkecil dibandingkan model lain dengan nilai MAE sebesar 11947.613069, nilai MAPE sebesar 0.212946, dan nilai RMSE sebesar 18865.036935. Dari hasil ini menjadi model dasar yang akan digunakan adalah Random Forest Regressor.**

*From [source](https://levelup.gitconnected.com/random-forest-regression-209c0f354c84), Random Forest Regression is a supervised learning algorithm that uses ensemble learning method for regression. Ensemble learning method is a technique that combines predictions from multiple machine learning algorithms to make a more accurate prediction than a single model.*


caps3.1.jpg

*The diagram above shows the structure of a Random Forest. You can notice that the trees run in parallel with no interaction amongst them. A Random Forest operates by constructing several decision trees during training time and outputting the mean of the classes as the prediction of all the trees. To get a better understanding of the Random Forest algorithm, lets walk through the steps:*

- Pick at random k data points from the training set.
- Build a decision tree associated to these k data points.
Choose the number N of trees you want to build and repeat steps 1 and 2.
- For a new data point, make each one of your N-tree trees predict the value of y for the data point in question and assign the new data point to the average across all of the predicted y values.

*A Random Forest Regression model is powerful and accurate. It usually performs great on many problems, including features with non-linear relationships. Disadvantages, however, include the following: there is no interpretability, overfitting may easily occur, we must choose the number of trees to include in the model.*

### **4.3.2. Hyperparameter Tunning**

Pada tahap ini akan melakukan tuning hiperparameter untuk mendapatkan parameter terbaik pada model terpilih (Random Forest Regressor). Random Forest Regressor memiliki banyak parameter. Jika kita mengubah semua parameter default dari model, akan ada kekhawatiran bahwa model akan overfitting. Sehingga parameter default yang diubah dari model hanya 3 sesuai dengan rekomendasi dari [sumber](https://www.analyticsvidhya.com/blog/2015/06/tuning-random-forest-model/), yaitu "max_depth" sebagai kedalaman pohon, "min_samples_leaf" memastikan bahwa node tersebut memiliki cukup sample untuk membuat generalisasi yang baik dari data dan meminimalisir variasi dari noise., dan "n_estimators" sebagai jumlah iterasi yang diinginkan.

```python
# Hyperparam space Random Forest Regresor
hyperparam_space_rf = {
    #'model__bootstrap': [True, False],
    'model__max_depth': [4,5,6,7,8],
    #'model__max_features': ['auto', 'sqrt'],
    'model__min_samples_leaf': [1, 2, 4],
    #'model__min_samples_split': [2, 5, 10],
    'model__n_estimators': [100,125,150,175,200]
}
```

```python
# Benchmark model dengan hyperparameter tuning
rf = RandomForestRegressor(random_state=2020)

# Membuat algorithm chains
estimator_rf = Pipeline([
        ('preprocessing', transformer),
        ('model', RandomForestRegresor)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=2020)

# Hyperparameter tuning
random_rf = GridSearchCV(
    estimator_rf, 
    param_grid = hyperparam_space_rf,
    cv = crossval, 
    scoring = 'neg_mean_absolute_percentage_error', 
    n_jobs = -1,
    verbose = 1
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_rf.fit(x_train,y_train)
```

```python
# Melihat hasil tuning dalam bentuk dataframe. Diurutkan berdasarkan RMSE, MAE & MAPE
pd.DataFrame(random_rf.cv_results_).sort_values('rank_test_score',ascending=True)
```

```python
# Cek skor dan params terbaik
print('Random Forest Regresor')
print('Best_score:', random_rf.best_score_)
print('Best_params:', random_rf.best_params_)
```

Dari hasil tuning hiperparameter, didapatkan perubahan berikut:

* model__min_samples_leaf : 2
* max_depth : 8
* n_estimators : 200

Kemudian, kita akan memasukkan perubahan parameter ke dalam model dan mencoba lagi pada data uji.

### ***Saya mencoba mengilustrasikan perkiraan bisnis model yang cocok untuk syarah.com dan akan membandingkan bagaimana perkembangan bisnis model sebelum dan juga setelah pembuatan machine learning.***

Berikut adalah contoh Business Model Canvas untuk sebuah perusahaan jual beli mobil bekas online di Saudi Arabia sebelum dan setelah pengaplikasian *machine learning* :

**Value Proposition**: Menawarkan platform jual beli mobil bekas online yang mudah digunakan, aman, dan memiliki koleksi mobil bekas terlengkap di Saudi Arabia. (***Setelah mengaplikasikan machine learning***
Menawarkan platform jual beli mobil bekas online yang menggunakan teknologi data dan machine learning untuk membantu konsumen membuat keputusan pembelian yang lebih baik dan memperoleh harga terbaik bagi mobil bekas mereka.)

**Customer Segments**: Konsumen yang ingin membeli atau menjual mobil bekas, baik individu maupun perusahaan.

**Channels**: Aplikasi mobile dan website yang dapat diakses melalui perangkat seluler dan komputer.

**Customer Relationships**: Interaksi dengan pelanggan melalui layanan pelanggan 24/7 dan fitur komunikasi seperti chat, email, dan telepon.

**Revenue Streams**: Pemasukan dari komisi pada setiap transaksi yang terjadi melalui platform, iklan dan sponsorship dari perusahaan otomotif, dan biaya tambahan untuk layanan tambahan seperti inspeksi mobil.(***Setelah mengaplikasikan machine learning***
Revenue bisa didapat juga dari komisitransaksi pengembangan usaha dengan jasa bengkel perbaikan bengkel yang bergabung di platform.)

**Key Resources**: Sumber daya manusia untuk mengelola platform dan menyediakan layanan pelanggan, teknologi untuk membuat platform mudah digunakan dan aman, dan data dan informasi pasar untuk membantu konsumen membuat keputusan.(***Setelah mengaplikasikan machine learning***
Penggunaan teknologi seperti machine learning dan big data untuk membuat platform lebih bermanfaat bagi konsumen)

**Key Partners**: Perusahaan otomotif, perusahaan inspeksi mobil, dan lembaga keuangan untuk membantu konsumen memproses transaksi.(***Setelah mengaplikasikan machine learning***
Perusahaan teknologi untuk memastikan platform memanfaatkan teknologi data dan machine learning dengan baik.)

**Key Activities**: Mengelola platform, menyediakan layanan pelanggan, memperbarui dan meningkatkan platform, memperoleh dan mempertahankan kepercayaan pelanggan.(***Setelah mengaplikasikan machine learning***
Mengumpulkan dan menganalisis data untuk memperbaiki pengalaman pelanggan.)

**Cost Structure**:: Biaya operasional seperti gaji karyawan, pemeliharaan teknologi, promosi dan pemasaran, dan biaya pemeliharaan data.

***Dengan machine learning syarah.com tidak hanya bisa menawarkan platform jual beli mobil bekas online yang mudah digunakan, aman, dan memiliki koleksi mobil bekas terlengkap, dan memberikan pengalaman belanja yang lebih baik bagi konsumen di Saudi Arabia. Namun bisa membantu konsumen membuat keputusan pembelian yang lebih baik dan memperoleh harga terbaik bagi mobil bekas mereka, perusahaan dapat meningkatkan pengalaman belanja bagi konsumen di sana.***

### **Untuk memperbaiki hasil prediksi, beberapa hal yang dapat dilakukan adalah:**

1. Menilai dan membagi error yang muncul ke dalam dua kategori, overestimasi dan underestimasi. Selanjutnya, memeriksa bagaimana error tersebut terkait dengan setiap variabel bebas. Ini akan membantu mengungkap aspek dan faktor-faktor yang menyebabkan tingginya error dan memberikan peluang untuk melakukan ulang modeling dengan menggunakan teknik perbaikan fitur yang berbeda.
2. Menambahkan sumber data agar model memiliki lebih banyak informasi untuk dipelajari, yaitu dengan menambahkan data teraktual tentang mobil bekas di Arab Saudi.
3. Menambahkan fitur yang memiliki korelasi dengan harga mobil, seperti kondisi mobil, garansi resmi, dan lain-lain.
4. Menentukan perbedaan antara mobil yang dikategorikan sebagai klasik dan mobil biasa yang digunakan untuk keperluan sehari-hari, karena hal ini akan mempengaruhi kaitan antara price dan mileage. Contohnya, sebuah mobil tahun 1980 akan memiliki harga yang lebih tinggi dibandingkan dengan mobil tahun 2007, tetapi membutuhkan pengetahuan domain untuk mengkategorikan mobil sebagai klasik.
5. Menyatakan opsi untuk membuat model tersendiri bagi mobil mewah atau fokus pada penjualan mobil biasa untuk kebutuhan sehari-hari, karena kehadiran data mobil mewah dapat memberikan dampak signifikan terhadap model.
6. Penambahan jumlah data perlu diberikan perhatian pada merek mobil yang sedikit jumlahnya dan harganya mahal. Sedikitnya jumlah data sangat mempengaruhi hasil model akhir yang diperoleh untuk sedikitnya merek mobil.
7. Pada masa depan, model ini perlu dibantu oleh model klasifikasi yang memiliki variabel target adalah penjualan yang "menguntungkan atau tidak" berdasarkan nilai harga mobil yang telah terjual.
